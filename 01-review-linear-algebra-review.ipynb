#### 1. Generate a Grid

```python
color_palette = ['#BE0E18', '#DD6317', '#E8D471', '#239176', '#20333F']

color_palette = np.repeat(color_palette, 5)

xvals = np.linspace(-2, 2, 5)
yvals = np.linspace(-2, 2, 5)
xygrid = np.column_stack([[x, y] for x in xvals for y in yvals])

plt.style.use('ggplot')
plt.scatter(xygrid[0], xygrid[1], s = 100, c = color_palette)
plt.title('The Original Grid', fontsize=10)
plt.axis('equal')
plt.show()
```

#### Shear

```python
a = np.array([[1,1],
              [0,1]])

# linear transformation
transformed = a@xygrid

plt.style.use('ggplot')
plt.scatter(transformed[0], transformed[1], s = 100, c = color_palette)
plt.title('Apply a Linear Transformation: Shear', fontsize=10)
plt.axis('equal')
plt.show()
```

#### Rotate

```python
a = np.array([[0,-1],
              [1, 0]])

# linear transformation
transformed = a@xygrid

plt.style.use('ggplot')
plt.scatter(transformed[0], transformed[1], s = 100, c = color_palette)
plt.title('Apply a Linear Transformation: Shear', fontsize=10)
plt.axis('equal')
plt.show()
```
